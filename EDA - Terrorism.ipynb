## Exploratory Data Analysis

Exploratory Data Analysis(EDA) is essentially a type of storytelling for statisticians. It is a way of visualizing, summarizing and interpreting the data that is hidden in rows and column format. Once EDA is complete and insights are drawn, its feature can be used for supervised and unsupervised machine learning modelling.

### Global Terrorism Database

The Global Terrorism Database is the most comprehensive unclassified database of terrorist attacks in the world. ... The GTD is an open-source database, which provides information on domestic and international terrorist attacks around the world since 1970, and now includes more than 200,000 events.

##### Visualizing the missing values

```python
plt.figure(figsize = (20, 5))
sns.heatmap(df.isnull(), cbar = False, yticklabels = False, cmap = "viridis")
```

In the above visualization, the number of missing values of each variable is denoted by the 'yellow' color occupancy in each variable.

```python
# Calculating the percentage of missing values of each variable

data_missing = df.isnull().sum().sort_values(ascending = False)

percent_missing = (df.isnull().sum()/df.isnull().count() * 100).sort_values(ascending = False)
missing_data = pd.concat([data_missing, percent_missing], axis = 1, keys = ['Data', '% Missing'])
missing_data
```

##### Dropping the variables with missing values more than 50%

```python
for i in df.columns:
    if missing_data.loc[i, '% Missing'] >= 50:
        df = df.drop(i, axis = 1)
```

##### Creating a dataframe with only the required columns for the dataset

```python
df1 = df[['iyear', 'imonth', 'iday','country_txt', 'provstate', 'region_txt', 'city', 'latitude', 'longitude', 'attacktype1_txt', 'nkill', 'nwound','target1', 'summary', 'gname', 'targtype1_txt', 'weaptype1_txt']]
```

##### Features of data

```python
print("Country with the most attacks: ", df1['country_txt'].value_counts().idxmax())
print("City with the most attacks: ", df1['city'].value_counts().index[1])
print("Region with the most attacks: ", df1['region_txt'].value_counts().idxmax())
print("Year with the most attacks: ", df1['iyear'].value_counts().idxmax())
print("Month with the most attacks: ", df1['imonth'].value_counts().idxmax())
print("A day with the most attacks: ", df1['iday'].value_counts().idxmax())
print("Group with the most attacks: ", df1['gname'].value_counts().index[1])
print("Most attack types: ", df1['attacktype1_txt'].value_counts().idxmax())
```

##### Terrorist Activities Each Year

```python
x_year = df1['iyear'].unique()
y_years = df1['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (25, 20))
sns.barplot(x = x_year, y = y_years, palette = 'mako')
plt.xticks(rotation = 45)
plt.title('Attack of Years')
plt.show()
```

##### Terrorist Activities by Region in each year

```python
pd.crosstab(df1['iyear'], df1['region_txt']).plot(kind = 'area', figsize = (15, 8))
plt.title('Terrorist Activities by Region in each Year')
plt.show()
```

##### Top Countries affected by Terror Attacks

```python
plt.subplots(figsize = (15, 6))
sns.barplot(df1['country_txt'].value_counts()[:15].index, df1['country_txt'].value_counts()[:15].values, palette = 'cubehelix')
plt.title('Top Countries Affected')
plt.xticks(rotation = 90)
plt.show()
```

##### Mapping the terrorist acts all over world in a certain year

```python
import folium
from folium.plugins import MarkerCluster
```

```python
filterYear = df1['iyear'] == 1970

filterData = df1[filterYear]

reqFilterData = filterData.loc[:, 'city':'longitude']

reqFilterData = reqFilterData.dropna()

reqFilterDataList = reqFilterData.values.tolist()
```

```python
map = folium.Map(location = [0, 30], tiles = 'openstreetmap', zoom_start = 5)

markerCluster = folium.plugins.MarkerCluster().add_to(map)
for point in range(0, len(reqFilterData)):
    folium.Marker(location = [reqFilterDataList[point][1], reqFilterDataList[point][2]],
                 popup = reqFilterDataList[point][0]).add_to(markerCluster)
map
```

##### Mapping the terrorist organization throughout the World

```python
df_terror_group = df1.dropna(subset = ['latitude', 'longitude'])
df_terror_group = df_terror_group.drop_duplicates(subset = ['country_txt', 'gname'])
terror_groups = df1['gname'].value_counts()[1:8].index.tolist()
df_terror_group = df_terror_group.loc[df_terror_group['gname'].isin(terror_groups)]
print(df_terror_group['gname'].unique())
```

```python
map = folium.Map(location = [20, 0], tiles = 'openstreetmap', zoom_start = 5)
markerCluster = folium.plugins.MarkerCluster().add_to(map)
for i in range(0, len(df_terror_group)):
    folium.Marker([df_terror_group.iloc[i]['latitude'], df_terror_group.iloc[i]['longitude']],
                 popup = 'Group:{}<br>Country:{}'.format(df_terror_group.iloc[i]['gname'],
                                                        df_terror_group.iloc[i]['country_txt'])).add_to(map)
map
```

##### Numer of killings in terror attack

```python
killData = df1.loc[:, 'nkill']
print('Number of people killed by terror attack :', int(sum(killData.dropna())))
```

##### Types of killings in terror attack

```python
attackData = df1.loc[:, 'attacktype1_txt']
typeKillData = pd.concat([attackData, killData], axis = 1)
```

```python
typeKillData = typeKillData.pivot_table(columns = 'attacktype1_txt', values = 'nkill', aggfunc = 'sum')
typeKillData
```

##### Plotting a pie chart to visualize the types of killings

```python
import plotly.express as px
labels = typeKillData.columns.tolist()
transpoze = typeKillData.T
values = typeKillData.values.tolist()
fig, ax = plt.subplots(figsize = (20, 20), subplot_kw = dict(aspect = 'equal'))
plt.pie(values, startangle = 90, autopct = '%.2f%%')
plt.title('Types of terror attacks.')
plt.legend(labels, loc = 'upper right', bbox_to_anchor = (1.3, 0.9), fontsize = 20)
plt.show()
```
